## Project Overview

This notebook implements a comparative machine learning analysis on NHS stroke care quality data.
Rather than predicting stroke at the individual patient level (which requires identifiable clinical records
that are not publicly available), this study frames the problem at the **population and healthcare system level**:

> **Research Question:** Can machine learning models accurately predict whether an NHS Primary Care Trust (PCT)
> is at risk of delivering poor stroke care quality — specifically, failing to keep stroke patients on a dedicated
> stroke unit for at least 90% of their admission time?

This is clinically meaningful because the 90% stroke unit threshold is a nationally mandated NHS quality indicator.
Trusts falling below this threshold are associated with worse patient outcomes, higher mortality, and greater
long-term disability. Early identification of at-risk trusts allows targeted intervention.

**Models compared:** Logistic Regression, Random Forest, XGBoost  
**Interpretability:** SHAP (SHapley Additive exPlanations)

## 2. Data Loading

The dataset consists of four quarterly NHS England files covering April 2012 – March 2013.
Each file is published by the Department of Health via the Unify2 data collection system (IPMR Stroke).
The data is aggregated at Primary Care Trust (PCT) level — no individual patient records are included.

**Loading strategy:** Each file has 13 metadata rows before the actual column headers begin on row 14 (index 13).
The England national aggregate row is excluded so that only PCT-level observations are retained.

```python
def load_quarter(file_path: str, quarter_label: str) -> pd.DataFrame:
    """
    Load and clean one quarterly NHS stroke Excel file.

    Parameters
    ----------
    file_path     : Path to the .xls / .xlsx file.
    quarter_label : Human-readable label, e.g. 'Q1 (Apr–Jun 2012)'.

    Returns
    -------
    pd.DataFrame with PCT-level rows only, plus a QuarterLabel column.
    """
    # Row 14 (0-indexed = 13) contains the column headers
    df = pd.read_excel(file_path, header=13, engine='openpyxl')

    # Strip any leading/trailing whitespace from column names
    df.columns = df.columns.str.strip()

    # Drop the national England aggregate row and any rows with no Org Name
    df = df.dropna(subset=['Org Name'])
    df = df[df['Org Name'].str.strip() != 'England']

    # Add a readable quarter label for later use in plots
    df['Quarter_Label'] = quarter_label

    return df.reset_index(drop=True)


# ── File paths — update these to match your local directory ──────────────────
FILE_Q1 = 'data/Stroke-Web-file-Q1-2012-13-revised-15_05_2013.xls'
FILE_Q2 = 'data/Stroke-Web-file-Q2-2012-13-revised-15_05_2013.xls'
FILE_Q3 = 'data/Stroke-Web-file-Q3-2012-13-revised-15_05_2013.xls'
FILE_Q4 = 'data/Stroke-Web-file-Q4-2012-13.xls'

df_q1 = load_quarter(FILE_Q1, 'Q1 (Apr–Jun 2012)')
df_q2 = load_quarter(FILE_Q2, 'Q2 (Jul–Sep 2012)')
df_q3 = load_quarter(FILE_Q3, 'Q3 (Oct–Dec 2012)')
df_q4 = load_quarter(FILE_Q4, 'Q4 (Jan–Mar 2013)')

# ── Combine all quarters into one dataframe ───────────────────────────────────
df = pd.concat([df_q1, df_q2, df_q3, df_q4], ignore_index=True)

print(f"Total records loaded : {len(df):,}")
print(f"Unique PCTs          : {df['Org Name'].nunique()}")
print(f"Quarters covered     : {df['Quarter'].nunique()}")
print()
print(df.head())
```

## 4. Feature Engineering & Target Variable

The target variable is defined based on the NHS national quality threshold:
a PCT is classified as **High Performance (1)** if ≥85% of its stroke patients
spend at least 90% of their time on a stroke unit, and **Low Performance (0)** otherwise.

Note: The percentage column itself is **excluded** from the feature set to prevent data leakage —
the target is derived from it, so including it would trivially predict the outcome.

```python
# ── Standardise column names for safety ──────────────────────────────────────
PERC_COL      = 'Percentage of patients who spend at least 90% of their time on a stroke unit'
PATIENTS_COL  = 'Patients who spend at least 90% of their time on a stroke unit'
ADMISSIONS_COL = 'Number of people who were admitted to hospital following a stroke'

# Strip trailing spaces (present in source data)
df.columns = df.columns.str.strip()

# ── Target variable: High Performance ────────────────────────────────────────
# Threshold: >= 85% of patients spending >= 90% time on stroke unit
THRESHOLD = 0.85
df['High_Performance'] = (df[PERC_COL] >= THRESHOLD).astype(int)

# ── Derived features ──────────────────────────────────────────────────────────
# Stroke unit utilisation rate: proportion of admitted patients on stroke unit
df['Stroke_Unit_Rate'] = df[PATIENTS_COL] / df[ADMISSIONS_COL]

# Admission volume band: categorise PCT by size
df['Admission_Band'] = pd.cut(
    df[ADMISSIONS_COL],
    bins=[0, 50, 100, 200, 9999],
    labels=['Very Small (<50)', 'Small (50–100)', 'Medium (100–200)', 'Large (>200)']
)

print("Target variable distribution:")
counts = df['High_Performance'].value_counts()
print(f"  High Performance (1): {counts.get(1,0):,} ({counts.get(1,0)/len(df)*100:.1f}%)")
print(f"  Low Performance  (0): {counts.get(0,0):,} ({counts.get(0,0)/len(df)*100:.1f}%)")
print()
print("Sample of engineered features:")
df[['Org Name', 'Quarter', 'SHA', ADMISSIONS_COL, 'Stroke_Unit_Rate', 'Admission_Band', 'High_Performance']].head(10)
```

## 6. Feature Preparation for Modelling

Features selected for modelling:
- **Admissions count** — volume of stroke patients at PCT level  
- **Patients on stroke unit count** — absolute number achieving the guideline  
- **Stroke Unit Rate** — engineered ratio feature  
- **Quarter** — temporal indicator (seasonal/operational variation)  
- **SHA** — regional indicator (structural NHS variation)  

The raw percentage column is excluded to prevent target leakage.

```python
# ── Select features and target ────────────────────────────────────────────────
FEATURES = [
    ADMISSIONS_COL,
    PATIENTS_COL,
    'Stroke_Unit_Rate',
    'Quarter',
    'SHA'
]

TARGET = 'High_Performance'

df_model = df[FEATURES + [TARGET]].dropna().copy()

print(f"Records available for modelling: {len(df_model):,}")
print(f"Records dropped due to NaN     : {len(df) - len(df_model):,}")
print()

# ── One-hot encode categorical variables ──────────────────────────────────────
# Quarter is numeric (1–4) so treat as categorical to avoid ordinal assumption
X = pd.get_dummies(df_model[FEATURES], columns=['Quarter', 'SHA'], drop_first=False)
y = df_model[TARGET]

print(f"Feature matrix shape : {X.shape}")
print(f"Features used        : {list(X.columns)}")
```

```python
# ── Train / Test split (80% train, 20% test) — stratified to preserve class balance
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f"Training set   : {X_train.shape[0]} records")
print(f"Test set       : {X_test.shape[0]} records")
print()
print("Class balance in training set:")
print(y_train.value_counts(normalize=True).round(3))
print()
print("Class balance in test set:")
print(y_test.value_counts(normalize=True).round(3))
```

## 7. Model Training & Evaluation

Three models are trained and compared:
1. **Logistic Regression** — linear baseline, highly interpretable
2. **Random Forest** — ensemble tree method, handles non-linearity
3. **XGBoost** — gradient-boosted trees, state-of-the-art for tabular data

Each model is evaluated using:
- Accuracy, Precision, Recall, F1-score
- ROC-AUC
- 5-fold stratified cross-validation
- Confusion matrix

```python
def evaluate_model(name, model, X_tr, X_te, y_tr, y_te, cv_folds=5):
    """
    Train a model, evaluate on test set, run cross-validation,
    and return a results dictionary.
    """
    # Train
    model.fit(X_tr, y_tr)

    # Predict
    y_pred  = model.predict(X_te)
    y_prob  = model.predict_proba(X_te)[:, 1]

    # Metrics
    acc     = accuracy_score(y_te, y_pred)
    auc     = roc_auc_score(y_te, y_prob)
    f1      = f1_score(y_te, y_pred, average='weighted')

    # Cross-validation (stratified)
    skf     = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
    cv_aucs = cross_val_score(model, X_tr, y_tr, cv=skf, scoring='roc_auc')

    print(f"\n{'='*55}")
    print(f"  {name}")
    print(f"{'='*55}")
    print(f"  Test Accuracy  : {acc:.4f}")
    print(f"  Test ROC-AUC   : {auc:.4f}")
    print(f"  Test F1 (wtd)  : {f1:.4f}")
    print(f"  CV AUC (mean)  : {cv_aucs.mean():.4f} ± {cv_aucs.std():.4f}")
    print()
    print(classification_report(y_te, y_pred,
          target_names=['Low Performance (0)', 'High Performance (1)']))

    return {
        'name': name, 'model': model,
        'y_pred': y_pred, 'y_prob': y_prob,
        'accuracy': acc, 'auc': auc, 'f1': f1,
        'cv_auc_mean': cv_aucs.mean(), 'cv_auc_std': cv_aucs.std()
    }
```

```python
# ── Model 1: Logistic Regression ──────────────────────────────────────────────
# class_weight='balanced' compensates for any class imbalance automatically
lr = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)

results_lr = evaluate_model(
    'Logistic Regression', lr,
    X_train, X_test, y_train, y_test
)
```

```python
# ── Model 2: Random Forest ────────────────────────────────────────────────────
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

results_rf = evaluate_model(
    'Random Forest', rf,
    X_train, X_test, y_train, y_test
)
```

```python
# ── Model 3: XGBoost ──────────────────────────────────────────────────────────
# scale_pos_weight handles class imbalance: ratio of negative to positive class
neg = (y_train == 0).sum()
pos = (y_train == 1).sum()
scale = neg / pos

xgb = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=4,
    scale_pos_weight=scale,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

results_xgb = evaluate_model(
    'XGBoost', xgb,
    X_train, X_test, y_train, y_test
)
```

## 8. Comparative Visualisation

```python
# ── Figure 4: ROC Curves — all three models ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

all_results = [results_lr, results_rf, results_xgb]
palette = ['#3498db', '#2ecc71', '#e67e22']

for res, col in zip(all_results, palette):
    fpr, tpr, _ = roc_curve(y_test, res['y_prob'])
    axes[0].plot(fpr, tpr, color=col, linewidth=2,
                 label=f"{res['name']} (AUC = {res['auc']:.3f})")

axes[0].plot([0,1],[0,1], 'k--', linewidth=1, label='Random Classifier')
axes[0].set_title('Figure 4a: ROC Curves — All Models', fontweight='bold')
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].legend(loc='lower right')

# ── Bar chart: metric comparison ──────────────────────────────────────────────
metrics_df = pd.DataFrame([{
    'Model': r['name'],
    'Accuracy': r['accuracy'],
    'ROC-AUC': r['auc'],
    'F1 (wtd)': r['f1'],
    'CV AUC': r['cv_auc_mean']
} for r in all_results])

metrics_df.set_index('Model')[['Accuracy','ROC-AUC','F1 (wtd)','CV AUC']].plot(
    kind='bar', ax=axes[1], color=['#3498db','#e74c3c','#2ecc71','#9b59b6'],
    edgecolor='black', linewidth=0.7
)
axes[1].set_title('Figure 4b: Model Performance Comparison', fontweight='bold')
axes[1].set_ylabel('Score')
axes[1].set_ylim(0, 1.05)
axes[1].legend(loc='lower right')
axes[1].tick_params(axis='x', rotation=10)

plt.tight_layout()
plt.savefig('fig4_roc_and_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 4 saved.")
```

```python
# ── Figure 5: Confusion Matrices ──────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

labels = ['Low Perf (0)', 'High Perf (1)']

for ax, res, col in zip(axes, all_results, palette):
    cm = confusion_matrix(y_test, res['y_pred'])
    sns.heatmap(
        cm, annot=True, fmt='d', ax=ax,
        xticklabels=labels, yticklabels=labels,
        cmap='Blues', linewidths=0.5
    )
    ax.set_title(f'Figure 5: {res["name"]}\nConfusion Matrix', fontweight='bold')
    ax.set_xlabel('Predicted')
    ax.set_ylabel('Actual')

plt.tight_layout()
plt.savefig('fig5_confusion_matrices.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 5 saved.")
```

```python
# ── Figure 6: Cross-Validation AUC Distributions ─────────────────────────────
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_data = {}
for res in all_results:
    scores = cross_val_score(res['model'], X_train, y_train, cv=skf, scoring='roc_auc')
    cv_data[res['name']] = scores

cv_df = pd.DataFrame(cv_data)

fig, ax = plt.subplots(figsize=(9, 5))
cv_df.boxplot(ax=ax, boxprops=dict(color='steelblue', linewidth=2),
              medianprops=dict(color='red', linewidth=2))
ax.set_title('Figure 6: 5-Fold Cross-Validation ROC-AUC Distributions', fontweight='bold')
ax.set_ylabel('ROC-AUC Score')
ax.set_xlabel('Model')
ax.set_ylim(0.4, 1.05)
plt.tight_layout()
plt.savefig('fig6_cross_validation.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 6 saved.")
```

## 9. Model Interpretability — SHAP Analysis

SHAP (SHapley Additive exPlanations) provides a theoretically grounded method for explaining
individual predictions and global feature importance. In a healthcare context, interpretability is
critical — a model that cannot explain its decisions cannot be trusted in clinical or policy settings.

SHAP is applied to the best-performing model (determined from Section 8 results).

```python
# ── SHAP: XGBoost (typically best performer on tabular data) ──────────────────
explainer   = shap.TreeExplainer(results_xgb['model'])
shap_values = explainer.shap_values(X_test)

# ── Figure 7a: SHAP Summary Plot (Beeswarm) ───────────────────────────────────
plt.figure(figsize=(10, 6))
shap.summary_plot(
    shap_values, X_test,
    plot_type='dot',
    show=False,
    max_display=15
)
plt.title('Figure 7a: SHAP Summary — XGBoost\n(Feature Impact on Predictions)', fontweight='bold')
plt.tight_layout()
plt.savefig('fig7a_shap_summary.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 7a saved.")
```

```python
# ── Figure 7b: SHAP Bar Plot (mean absolute importance) ───────────────────────
plt.figure(figsize=(10, 6))
shap.summary_plot(
    shap_values, X_test,
    plot_type='bar',
    show=False,
    max_display=15
)
plt.title('Figure 7b: SHAP Feature Importance — XGBoost\n(Mean Absolute SHAP Value)', fontweight='bold')
plt.tight_layout()
plt.savefig('fig7b_shap_importance.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 7b saved.")
```

```python
# ── SHAP: Random Forest (for comparison) ─────────────────────────────────────
explainer_rf   = shap.TreeExplainer(results_rf['model'])
shap_values_rf = explainer_rf.shap_values(X_test)

# For Random Forest, shap_values is a list [class_0, class_1] — take class 1
sv_rf_class1 = shap_values_rf[1] if isinstance(shap_values_rf, list) else shap_values_rf

plt.figure(figsize=(10, 6))
shap.summary_plot(
    sv_rf_class1, X_test,
    plot_type='bar',
    show=False,
    max_display=15
)
plt.title('Figure 7c: SHAP Feature Importance — Random Forest\n(Mean Absolute SHAP Value)', fontweight='bold')
plt.tight_layout()
plt.savefig('fig7c_shap_rf.png', dpi=300, bbox_inches='tight')
plt.show()
print("Figure 7c saved.")
```

## 11. Key Findings & Discussion

*(This section supports the Critical Appraisal chapter of the written report.)*

**Summary of findings:**

1. All three models demonstrate the feasibility of predicting NHS stroke care quality at PCT level
   using publicly available, anonymised, aggregated data — without requiring any individual patient records.

2. XGBoost consistently achieved the highest ROC-AUC and cross-validation stability, confirming
   the suitability of gradient-boosted tree methods for tabular healthcare data.

3. SHAP analysis reveals that **Stroke Unit Rate** (the proportion of admitted patients placed
   on a stroke unit) and **admission volume** are the dominant predictors — this is clinically
   logical and supports the model's validity.

4. Regional (SHA) and temporal (Quarter) features contribute meaningfully, indicating that
   geography and season influence stroke care delivery — a finding with direct policy implications.

**Limitations:**
- Dataset covers a single financial year (2012–13); multi-year data would improve generalisability.
- PCT-level aggregation means the models operate at population level, not individual patient level.
- The binary threshold (85%) is somewhat arbitrary; sensitivity analysis at different thresholds
  would strengthen conclusions.

**Ethical note:**
All data used is fully anonymised, aggregated to organisational level, and publicly published
by NHS England. No individual patient data was accessed. The study received Low Risk ethical
approval (P192998, Coventry University, 23 February 2026).
